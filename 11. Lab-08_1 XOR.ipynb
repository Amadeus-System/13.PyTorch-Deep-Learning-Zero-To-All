# Lab 9 : XOR

```python
import torch
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# For reproducibility
torch.manual_seed(777)
if device == 'cuda': # 만약 cuda 환경이면
    torch.cuda.manual_seed_all(777) # cuda 안의 모든 seed를 재설정
```

```python
# XOR 데이터셋
X = torch.FloatTensor([[0, 0],
                       [0, 1],
                       [1, 0],
                       [1, 1]]).to(device)
Y = torch.FloatTensor([[0],
                       [1],
                       [1],
                       [0]]).to(device)
```

```python
# nn layers
linear = torch.nn.Linear(2, 1, bias = True)
sigmoid = torch.nn.Sigmoid()
```

```python
# model
model = torch.nn.Sequential(linear, sigmoid).to(device) # 단층 신경망 구성
```

```python
# define cost/loss & optimizer
criterion = torch.nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr = 1)
```

```python
for step in range(10001):

    optimizer.zero_grad()
    hypothesis = model(X)

    # cost/loss function
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()

    if step % 100 == 0:
        print(step, cost.item()) # cost가 잘 감소하지 않는다. 학습이 잘 되지 않는다는 뜻!
```

```python
# Accuracy computation
# True if hypothesis > 0.5 else False
with torch.no_grad():
    hypothesis = model(X)
    predicted = (hypothesis > 0.5).float() # True -> 1, False -> 0
    accuracy = (predicted == Y).float().mean()
    print('\nHypothesis:', hypothesis.detach().cpu().numpy(),
          '\nCorrect:', predicted.detach().cpu().numpy(),
          '\nAccuracy:', accuracy.item())
```

정확도가 50%가 되어 단층 신경망으로는 XOR 문제를 해결할 수 없다.
